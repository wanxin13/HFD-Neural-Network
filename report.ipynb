```python
def getReturns(data, N, T):
    re = []
    for t in range(T):
        for i in range(N-1):
            re.append(data['price'][N*t+i+1]/data['price'][N*t+i])
    ret = np.log(re)
    ret = ret.reshape((T,N-1))
    return ret
    
N = 78
T = 2769
ret = getReturns(data, 78, 2769)
print(ret)
```

```python
def getBV(ret,N,T):
    bv =[]
    for t in range(T):
        bvt = 0.0
        for i in range(1,N-1):
            bvt = bvt + abs(ret[t][i])*abs(ret[t][i-1])
        bvt =(math.pi/2)*bvt
        bv.append(bvt)
    return bv

BV = getBV(ret,N,T)
print(BV)
```

```python
def getTOD(ret,N,T):
    tod =[]
    b = np.empty((78), dtype='float_')
    for i in range(1,N-1):
        bs = 0
        for t in range(T):
            bs = bs + abs(ret[t][i]*ret[t][i-1])
        bs = bs/T
        b[i] = bs
    b[0]=b[1]
    for i in range(N-1):
        tod.append(((N-1)*b[i])/np.sum(b))
    return tod

tod = getTOD(ret,N,T)
print(tod)
```

```python
def separateReturns(ret,alpha=5):
    cut = np.empty((T, N-1), dtype='float_')
    rc = np.zeros((T, N-1), dtype='float_')
    rd = np.zeros((T, N-1), dtype='float_')
    nd = 0
    # compute jump threshold
    for t in range(T):
        for i in range(N-1):
            cut[t][i] = alpha*((1/(N-1))**(0.49))*np.sqrt(tod[i]*BV[t])
    # separate diffusive from jump returns
    # return diffusive and jump returns
    for t in range(T):
        for i in range(N-1):
            if abs(ret[t][i])<= cut[t][i]:
                rc[t][i]= ret[t][i]
            else:
                rd[t][i]= ret[t][i]
                nd = nd+1
    return rc, rd, nd

rc, rd, nd = separateReturns(ret)
print(rc)
print(rd)
print(nd)
```

```python
def getQIV(rc):
    QIV = np.empty((T),dtype='float_')
    for t in range(T):
        rcsum = np.sum(rc[t]**4)
        QIV[t]= ((N-1)/3)*rcsum
    return QIV

QIV = getQIV(rc)
print(QIV)
```

```python
def getRVWeek(RV):
    # compute all RVw
    RVw = np.zeros((T),dtype='float_')
    for t in range(21,T):
        RVw[t] = np.sum(RV[t-4:t+1])/5
    return RVw
        
def getRVMonth(RV):
    # compute all RVm
    RVm = np.zeros((T),dtype='float_')
    for t in range(21,T):
        RVm[t] = np.sum(RV[t-21:t+1])/22
    return RVm


RVw = getRVWeek(RV)
RVm = getRVMonth(RV)

model_input = np.ones((T-22+1, 5), dtype='float_')
model_input[:, 1] = RV[21:]
model_input[:, 2] = RVw[21:]
model_input[:, 3] = RVm[21:]
model_input[:, 4] = np.sqrt(QIV[21:])*RV[21:]
print(model_input.shape)
columns = model_input.shape[1]
print(model_input)
```

Exercise2B

```python
def OLS(Y, X):
    # return beta estimate from OLS
    # np.linalg.inv computes the inverse of a matrix
    # remember that in numpy * is elementwise multiplication
    # the symbol for matrix multiplication is @, as in X.T@X
    # X.T is the tranpose of X
    beta = np.linalg.inv(X.T@X)@X.T@Y
    return beta
```

```python
def forecastError(Y, X, Y_next, X_next):
    # computes the forecast error
    # estimate beta
    beta = OLS(Y,X)
    Y_forecast = X_next@beta
    return (Y_next - Y_forecast)**2
```

```python
error = {'AR1': 0.0, 'HAR1': 0.0, 'ARQ1': 0.0, 'HARQ1': 0.0, 'NoChange': 0.0}
window_size = 1000
# rolling window + 1-step-ahead forecast
for S in range(window_size,T-22):
    window_x = slice(S - window_size, S)
    window_y = slice(S - window_size + 1, S + 1)
    # AR(1)
    error['AR1'] += forecastError(model_input[window_y, 1],
                                  model_input[window_x, 0:2],
                                  model_input[S+1, 1],
                                  model_input[S,0:2])
    # HAR(1)
    error['HAR1'] += forecastError(model_input[window_y,1],
                                   model_input[window_x, 0:4],
                                   model_input[S+1,1],
                                   model_input[S,0:4])
    # ARQ(1)
    Y_forecast = model_input[S,[0,1,4]] @ OLS(model_input[window_y,1],model_input[window_x,[0,1,4]])
    if Y_forecast > np.max(model_input[window_y,1]) or Y_forecast < np.min(model_input[window_y,1]) :
        Y_forecast = np.mean(model_input[window_y,1])
        error['ARQ1'] += (model_input[S+1,1] - Y_forecast)**2 
    else:
        error['ARQ1'] += forecastError(model_input[window_y,1],
                                   model_input[window_x,[0,1,4]],
                                   model_input[S+1,1],
                                   model_input[S,[0,1,4]])
    
    # HARQ(1)
    Y_forecast = model_input[S,:] @ OLS(model_input[window_y,1],model_input[window_x,:])
    if Y_forecast > np.max(model_input[window_y,1]) or Y_forecast < np.min(model_input[window_y,1]) :
        Y_forecast = np.mean(model_input[window_y,1])
        error['HARQ1'] += (model_input[S+1,1] - Y_forecast)**2 
    else:
        error['HARQ1'] += forecastError(model_input[window_y,1],
                                    model_input[window_x,:],
                                    model_input[S+1,1],
                                    model_input[S,:])
    # No Change
    error['NoChange'] += (model_input[S+1,1]-model_input[S,1])**2
print(error)
```

```python
# compute and print MSE
total = len(model_input) - window_size 
print("{:^10}|{:^10}".format('Model','MSE'))
print(f"{'-'*21}")
for model, value in error.items():
    print(model,value/total)
print(f"{'-'*21}")
```

```python
import tensorflow as tf
# get default graph
# create model input (x) and model output (x)
x = tf.placeholder(tf.float32, [None, 5])
y = tf.placeholder(tf.float32, [None])
# create hidden layer
W1 = tf.Variable(tf.random_normal([5,10]))
S1 = tf.nn.relu(tf.matmul(x,W1))
# create last layer
W2 = tf.Variable(tf.random_normal([10,1]))
y_model = tf.matmul(S1,W2)
```

Exercise3B

```python
# create loss function
loss = tf.reduce_mean(tf.squared_difference(y_model,y))
# create optimizer
train_one_step = tf.train.AdamOptimizer(learning_rate = 0.005).minimize(loss)
```

Exercise3C

Notion: I set epochs=20000 since I found MSE remained quite stable aftesr 15000.

```python
def forecastNN(S, window_size=1000, epochs=20000, columns=5):
    # determine window size
    window_x = slice(S - window_size, S)
    window_y = slice(S - window_size +1, S+1)
    with tf.Session() as session:
        # initialize variables
        session.run(tf.global_variables_initializer())
        print(f'\n\nSTEP | MSE')
        for i in range(epochs):
            # do 1 step on gradient descent
            [_,MSE]= session.run([train_one_step, loss],feed_dict={
            x: model_input[window_x,:],
            y: model_input[window_y,1]
        })
            # print MSE every 500 epochs
            if i % 500 == 0:
                print(f'{i:^5}|{MSE:^4}')
        # model is done training, use it to forecast
        y_forecast = session.run(y_model, feed_dict={
            x: [model_input[S,:]]
        })
        # add a sanity filter
        if y_forecast > np.max(model_input[window_y,1]) or y_forecast < np.min(model_input[window_y,1]) :
            y_forecast = np.mean(model_input[window_y,1])
        return float(y_forecast)
```
